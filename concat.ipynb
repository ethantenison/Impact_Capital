```python
import pandas as pd 
from IPython.display import display

```

# PFS

```python
pf2017 = pd.read_csv('data/pf/2017_objects.csv')
pf2018 = pd.read_csv('data/pf/2018_objects.csv')
pf2019 = pd.read_csv('data/pf/2019_objects.csv')

pfs = pd.concat([pf2017,pf2018, pf2019])
pfs2017 = pfs['TaxYr'] == 2017
pf2017 = pfs[pfs2017]
```

```python
pd.options.display.max_columns = None
pf2017.to_csv("data/PF_2017.csv",index=False)
display(pf2017)
```

#  990's 

```python
reg2017 = pd.read_csv('data/2017_objects.csv')
reg2018 = pd.read_csv('data/2018_objects.csv')
reg2019 = pd.read_csv('data/2019_objects.csv')

regs = pd.concat([reg2017,reg2018, reg2019])
regs2017 = regs['TaxYr'] == 2017
reg2017 = regs[regs2017]

pd.options.display.max_columns = None
reg2017.to_csv("data/Form_990_2017.csv",index=False)
display(reg2017)
```

# EZ
